```python
# preparing upload Git hub
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import accuracy_score


df = pd.read_csv('/home/jinho/Desktop/work/jinho/Learning_ML/dataset/CICIDS2017/MachineLearningCVE/Wednesday-workingHours.pcap_ISCX.csv',low_memory=False)
# # df.info()
# df.shape

#Remove nan, Infinity values
df = df[~df.isin(['Infinity','NaN',np.nan, np.inf, -np.inf]).any(1)]

# # to make sure 
# df.shape

print("......Load Data......")
array = df.values

#befor Normalize
X = array[:, 1:-1]
# Normalize
scaler = preprocessing.MinMaxScaler()
X = scaler.fit_transform(X)


#before Label Encoding
Y = array[:, -1]
# Label Encoding
le = preprocessing.LabelEncoder()
Y = le.fit_transform(Y)


x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=1) 

# find best K_value
for k_value in range(1,14,2) :
    print("K is  =",k_value)
    knn = KNeighborsClassifier(n_neighbors=k_value ,weights='distance')  
    knn.fit(x_train, y_train) 
    y_pred = knn.predict(x_test)
    y_predict = knn.predict(x_train)

    accuracy = accuracy_score(y_test, y_pred)
    print('Accuracy: %f' % accuracy)
    print(precision_recall_fscore_support(y_test, y_pred, average='macro'))
    print("Train")
    accuracy2 = accuracy_score(y_train, y_predict)
    print('Train Accuracy: %f' % accuracy2)
    print(precision_recall_fscore_support(y_train, y_predict, average='macro'))
    print()
    
# K is  = 5
# Accuracy: 0.999031
# (0.9935887508635943, 0.9132501742648745, 0.9395156252687792, None)
# Train
# Train Accuracy: 0.999929
# (0.9999686055184484, 0.9999429367980547, 0.9999557685141626, None)
```
